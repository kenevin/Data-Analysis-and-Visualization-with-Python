Lambda functions
Look at the examples below.
There is no difference between these two.
Traditional function must be named. A lambda function does not have to be named which makes it convenient to pass to another function.

```python
def squareIt(x):
    return x**2
```

```python
squareItAgain = (lambda x:x**2)(4)
print squareItAgain
```

```python
def squareIt(x):
    return x**2
```

```python
def cubeIt(x):
    return x**3
```

```python
def powerOfFour(x):
    return x**4
```

```python
def powerOf(p):
    return lambda x:x**p # This will return a function
```

```python
powerOf(3) # This is called a function factory
```

```python
cubed = powerOf(3)
print cubed
```

```python
cubed(2) # And this is now called closure
```

```python
def stat(x):
    return pd.Series({'count': len(x), 'sum': np.sum(x),
                      'min': np.min(x), 'max': np.max(x),
                      'mean': np.mean(x), 'std': np.std(x)})

tips.groupby('sex')['tip'].apply(stat)
```

```python
tips.groupby('sex')['tip'].agg(['count', 'sum', 'min', 'max', 'mean', 'std'])
```

```python
tips.groupby('sex')['tip'].agg(['count', 'sum', 'min', 'max', 'mean', \
                               ('StandardDeviation', 'std')])
```

```python
colFun = {'tip': ['mean', 'std', 'max'],
         'total_bill': ['mean'],
         'size': ['sum', 'mean']}
tips.groupby('sex').agg(colFun)
```

```python
mostFrequent = lambda x:x.value_counts().index[0]
tips.groupby('sex').agg({'total_bill': np.mean, 'time': mostFrequent})
```

```python
tips.groupby('sex', as_index = False).agg(colFun)
```
